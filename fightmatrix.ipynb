```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select

import time
import pandas as pd

chrome_options = webdriver.ChromeOptions()
prefs = {"profile.managed_default_content_settings.images": 2}
chrome_options.add_experimental_option("prefs", prefs)
chrome_options.add_argument('--headless') 
#chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument('--disable-gpu')
```

```python
def select_dropdown(dropdown,index):
    select = Select(dropdown)
    select.select_by_index(index)

def current_page(new_url):
    page_index = new_url.find("Page=")
    page_string = new_url[page_index + len("Page="):]
    page_value = page_string.split("&")[0].rstrip("/")  
    return page_value

def scrap_data(divi_name,date):
    data = []
    ranks = driver.find_elements(By.CLASS_NAME, "tdRank")
    row = []
    for rank in ranks:
        if len(row) ==0:
            row.append(date)
            row.append(divi_name)
        row.append(rank.text)
        if len(row) ==5:
            data.append(row)
            row =[]

    ranks = driver.find_elements(By.CLASS_NAME, "tdRankAlt")
    row = []
    for rank in ranks:
        if len(row) ==0:
            row.append(date)
            row.append(divi_name)
        row.append(rank.text)
        if len(row) ==5:
            data.append(row)
            row =[]
    return data
```

```python
the_link= input("what is the link?")
divi_name = input("division name?")
counter_s = input("loop how many times?")
counter = int(counter_s)
current_date = input("current date?")
```

```python
#128
driver =webdriver.Chrome()
#driver.get("https://www.fightmatrix.com/historical-mma-rankings/ranking-snapshots/?Issue=332&Division=1")
#divi_name="heavyweight"

driver.get(the_link)
master_data = []

for j in range(1,counter):
    if j == 1:
        print("yo")
    else:
        time.sleep(1.5)
        #select new date, and grab date
        dropdown = driver.find_element(By.NAME,"Issue")
        select_dropdown(dropdown,j)
        dropdown_ele = driver.find_element(By.NAME,"Issue")
        selected_option_element = Select(dropdown_ele).first_selected_option
        current_date = selected_option_element.text
        print(current_date)
        

#Step1) Go to Last Page
    try:
        time.sleep(1.5)
        last_page = driver.find_element(By.XPATH,'//*[@class="pager"]/tbody/tr/td/a[2]')  
        last_page.click()
        scraped_data = scrap_data(divi_name)
        master_data.extend(scraped_data)
    except:
        print("no last page")

    #Step2) Find how many pages 
    new_url = driver.current_url
    pre_page = current_page(new_url)
    page = int(pre_page)
    page -= 1

    #Step3)Loop through all the pages, and grab data
    for i in range(page):
        time.sleep(1.5)
        next_page = driver.find_element(By.XPATH,'//*[@class="pager"]/tbody/tr/td/a[2]')
        next_page.click()

        time.sleep(1)
        scraped_data = scrap_data(divi_name,current_date)
        master_data.extend(scraped_data)
        

        
#

print(master_data)
```
